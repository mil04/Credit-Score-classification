<h1>TPOT - BEST ACCURACY </h1>

```python
from tpot import TPOTClassifier
import pandas as pd
import sklearn.metrics
from sklearn.model_selection import train_test_split


data_all = pd.read_csv('../data/data.csv')

#map CAT_GAMBLING to 1, 2, 3
data_all['CAT_GAMBLING'] = data_all['CAT_GAMBLING'].map({'No': 1, 'Low': 2, 'High': 3})

train, test = train_test_split(data_all, test_size=0.2, random_state=42)

X_train = train.drop(['CREDIT_SCORE', 'DEFAULT', 'CUST_ID'], axis=1)
y_train = train['DEFAULT']

X_test = test.drop(['CREDIT_SCORE', 'DEFAULT', 'CUST_ID'], axis=1)
y_test = test['DEFAULT']

tpot = TPOTClassifier(generations=30, population_size=40, verbosity=2, random_state=42, scoring = 'accuracy')
tpot.fit(X_train, y_train)
print(tpot.score(X_test, y_test))
```

<h1>TPOT - BEST RECALL 1 </h1>

```python
from tpot import TPOTClassifier
import pandas as pd
from sklearn.metrics import make_scorer, recall_score
from sklearn.model_selection import train_test_split

data_all = pd.read_csv('Project1/data/data.csv')

train, test = train_test_split(data_all, test_size=0.2, random_state=42)

X_train = train.drop(['CREDIT_SCORE', 'DEFAULT', 'CUST_ID', 'CAT_GAMBLING'], axis=1)
y_train = train['DEFAULT']
X_test = test.drop(['CREDIT_SCORE', 'DEFAULT', 'CUST_ID', 'CAT_GAMBLING'], axis=1)
y_test = test['DEFAULT']

scorer = make_scorer(recall_score, pos_label=1)

tpot = TPOTClassifier(generations=100, population_size=40, verbosity=2, random_state=42, scoring=scorer)

tpot.fit(X_train, y_train)

print("Best recall score (class 1) found:", tpot.score(X_test, y_test))
```
